```python

# Define the date patterns to match
date_patterns = [
    r"\d{1,2}/\d{1,2}/\d{4}",  # d/m/yyyy or dd/m/yyyy or dd/mm/yyyy
    r"\d{1,2}\s(?:JANUARI|FEBRUARI|MAC|APRIL|MEI|JUN|JULAI|OGOS|SEPTEMBER|OKTOBER|NOVEMBER|DISEMBER)\s\d{4}",  # d MONTH yyyy
    r"\d{1,2}-\w{3}-\d{2}",  # dd-Dec-yy
    r"\d{1,2}-\w{3}-\d{2}",  # dd-Dec-yy
]

# Filter the DataFrame to keep only valid date values
valid_dates = []
for value in df["TARIKH KEPUTUSAN UJIAN"]:
    if isinstance(value, str):
        for pattern in date_patterns:
            match = re.match(pattern, value)
            if match:
                valid_dates.append(match.group())
                break
        else:
            valid_dates.append(None)
    else:
        valid_dates.append(None)

df["TARIKH KEPUTUSAN UJIAN"] = valid_dates

# Drop rows with non-date values
df.dropna(subset=["TARIKH KEPUTUSAN UJIAN"], inplace=True)

# Convert the column to datetime format
df["TARIKH KEPUTUSAN UJIAN"] = pd.to_datetime(df["TARIKH KEPUTUSAN UJIAN"], format="%d/%m/%Y", errors="coerce")

# Reset the index if needed
df.reset_index(drop=True, inplace=True)

# Display the remaining date values
print(df["TARIKH KEPUTUSAN UJIAN"])
```
