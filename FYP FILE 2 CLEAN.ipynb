```python
#clean simptom pesakit column

# Define a custom function to concatenate the symptom columns
def combine_symptoms(row):
    # Create a list of symptoms that are "yes" for this row
    symptoms = []
    if row["SAKIT TEKAK: YA / TIDAK"] == "YA":
        symptoms.append("Sakit Tekak")
    if row["DEMAM : YA / TIDAK"] == "YA":
        symptoms.append("Demam")
    if row["SELESEMA : YA / TIDAK"] == "YA":
        symptoms.append("Selsema")
    if row["HILANG DERIA RASA : YA / TIDAK"] == "YA":
        symptoms.append("Hilang Deria Rasa")
    if row["HILANG DERIA BAU : YA / TIDAK"] == "YA":
        symptoms.append("Hilang Deria Bau")
    if row["BATUK : YA / TIDAK"] == "YA":
        symptoms.append("Batuk")
    if row["SESAK NAFAS : YA / TIDAK"] == "YA":
        symptoms.append("Sesak Nafas")
    if row["CIRIT BIRIT : YA / TIDAK"] == "YA":
        symptoms.append("Cirit Birit")
    
    # Add comorbidities if they are present
    if row["Komorbid"] == "YA":
        symptoms.append("Komorbid")
    
    # Check if the list of symptoms is empty, then add "None" to the list
    if not symptoms:
        symptoms.append("Tiada")
    
    # Join the list of symptoms into a single string
    return ", ".join(symptoms)

# Apply the function to create a new "Symptoms" column
df2["Simptom Pesakit"] = df2.apply(combine_symptoms, axis=1)

# Drop the individual symptom columns
df2 = df2.drop(["SAKIT TEKAK: YA / TIDAK", "DEMAM : YA / TIDAK", "SELESEMA : YA / TIDAK", 
              "HILANG DERIA RASA : YA / TIDAK", "HILANG DERIA BAU : YA / TIDAK", 
              "BATUK : YA / TIDAK", "SESAK NAFAS : YA / TIDAK", "CIRIT BIRIT : YA / TIDAK", 
              "Komorbid"], axis=1)
```

```python
df2['Simptom Pesakit'].value_counts()
```

```python
def categorize_simptom(simptom):
    if simptom == 'Tiada' or len(simptom.split()) <= 2:
        return 'Simptom Biasa'
    elif len(simptom.split()) <= 4:
        return 'Simptom Sederhana'
    else:
        return 'Simptom Kuat'

df2['Simptom Pesakit'] = df2['Simptom Pesakit'].apply(categorize_simptom)
```
