```python
import pandas as pd
import numpy as np
from dateutil.parser import parse
import re
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#FILE 4

df4 = pd.read_csv("Covidfile4sort.csv")
df4


```

```python

# Define the date patterns to match
date_patterns = [
    r"\d{1,2}/\d{1,2}/\d{4}",  # d/m/yyyy or dd/m/yyyy or dd/mm/yyyy
    r"\d{1,2}\s(?:JANUARI|FEBRUARI|MAC|APRIL|MEI|JUN|JULAI|OGOS|SEPTEMBER|OKTOBER|NOVEMBER|DISEMBER)\s\d{4}",  # d MONTH yyyy
    r"\d{1,2}-\w{3}-\d{2}",  # dd-Dec-yy
    r"\d{1,2}-\w{3}-\d{2}",  # dd-Dec-yy
]

# Filter the DataFrame to keep only valid date values
valid_dates = []
for value in df4["Tarikh Isytihar"]:
    if isinstance(value, str):
        for pattern in date_patterns:
            match = re.match(pattern, value)
            if match:
                valid_dates.append(match.group())
                break
        else:
            valid_dates.append(None)
    else:
        valid_dates.append(None)

df4["Tarikh Isytihar"] = valid_dates

# Drop rows with non-date values
df4.dropna(subset=["Tarikh Isytihar"], inplace=True)

# Convert the column to datetime format
df4["Tarikh Isytihar"] = pd.to_datetime(df4["Tarikh Isytihar"], format="%d/%m/%Y", errors="coerce")

# Reset the index if needed
df4.reset_index(drop=True, inplace=True)

# Display the remaining date values
print(df4["Tarikh Isytihar"])
```

```python
# Filter out all rows with a date from 2020
df4 = df4[df4['Tarikh Isytihar'].dt.year != 2020]

df4
```

```python
# Filter out all rows with a date from 2020
df4 = df4[df4['Tarikh Isytihar'].dt.year != 2022]

df4
```

```python
df4.to_csv('FILE4CLEANED.csv', index=False)
```
